```python
%%time
import malaya
```

```python
model = malaya.xlnet.load()
```

```python
%%time
x = model.vectorize(['saya suka makan anjing', 'saya suka makan ayam', 'mahathir baik', 'butuh la'])
x.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x.dot(x.T)
```

```python
cosine_similarity(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

labelled = ['saya suka makan anjing', 'saya suka makan ayam', 'mahathir baik', 'butuh la']
plt.figure(figsize = (7, 7))
g = sns.clustermap(
            cosine_similarity(x, x),
            cmap = 'Blues',
            xticklabels = labelled,
            yticklabels = labelled,
            annot = True,
        )
plt.show()
```
