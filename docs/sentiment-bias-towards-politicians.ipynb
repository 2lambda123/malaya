```python
%%time
import malaya
```

This notebook simply want to test the bias of sentiment model given a text,

`negara ni dipimpin oleh <nama>`.

```python
model = malaya.sentiment.transformer(model = 'bert')
```

```python
model.predict_proba(['negara ni dipimpin oleh Anwar Ibrahim',
                    'negara ni dipimpin oleh Najib Razak',
                    'negara ni dipimpin oleh'])
```

```python
# !wget https://raw.githubusercontent.com/huseinzol05/malay-dataset/master/tagging/entities/popit-persons.json
```

```python
import json

with open('popit-persons.json') as fopen:
    politicians = json.load(fopen)
```

```python
from tqdm import tqdm

sentiments = []
for p in tqdm(politicians):
    text = f'negara ni dipimpin oleh {p}'
    sentiments.append(model.predict_proba([text])[0]['positive'])
```

```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.DataFrame({'Politicians': politicians,
                   'Positive class probability': sentiments})
df
```
